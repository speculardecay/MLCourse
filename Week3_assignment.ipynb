```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"D:\ML\Intro to ML Course\Week3\Assignment\Titanic-Dataset.csv")
df.head()
```

Pandas Plotting

```python
# Age of all passengers
df.plot(kind="scatter", x="PassengerId", y="Age")
```

```python
# Age of survivors
dfTemp = df
for x in dfTemp.index:
    if dfTemp.loc[x, "Survived"] == 0:
        dfTemp.drop(x, inplace=True)
      
dfTemp.plot(kind="scatter", x="PassengerId", y="Age")
```

```python
# Age of surviving males
for x in dfTemp.index:
    if dfTemp.loc[x, "Sex"] == "female":
        dfTemp.drop(x, inplace=True)
      
dfTemp.plot(kind="scatter", x="PassengerId", y="Age")
```

```python
df = pd.read_csv(r"D:\ML\Intro to ML Course\Week3\Assignment\Titanic-Dataset.csv")
df.head()
```

```python
# Fare by Age
sorteddf = df.sort_values(by="Age")
sorteddf.plot(kind="line", x="Age", y="Fare")
```

```python
# Number of siblings/spouses by Age
sorteddf = df.sort_values(by="Age")
sorteddf.plot(kind="line", x="Age", y="SibSp")
```

```python
# Number of siblings/spouses by Age
sorteddf = df.sort_values(by="Age")
sorteddf1 = sorteddf["SibSp"]
sorteddf2 = sorteddf["Parch"]

plt.plot(sorteddf["Age"],sorteddf1,label="Siblings/Spouses (SibSp)")
plt.plot(sorteddf["Age"],sorteddf2, label="Parents/Children (Parch)")
plt.legend()
plt.title("Number of Siblings/Spouses and Parents/Children by Age")
plt.xlabel("Age")
plt.ylabel("Count")
```

Seaborn Plotting

```python
df = pd.read_csv(r"D:\ML\Intro to ML Course\Week3\Assignment\Titanic-Dataset.csv")
df.head()
```

```python
# Age by class
sns.catplot(
    data=df,
    x="Age", hue="Pclass", kind="count"
)
```

```python
# Age of survivors
sns.displot(
    data=df,
    x="Age", hue="Survived", kind="ecdf"
)
```

```python
# Age by Passenger Class
sns.displot(
    data=df,
    x="Age", col="Pclass", kde=True
)
```

```python
# Surviving sex
sns.displot(
    data=df,
    x="Sex", col="Survived"
)
```

```python
# Surviving sex age
sns.catplot(
    data=df,
    x="Sex", y="Age", hue="Survived", kind="violin"
)
```
