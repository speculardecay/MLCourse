```python
df.drop(["fueltype","symboling","aspiration","doornumber","carbody","carheight","enginetype","fuelsystem","compressionratio","peakrpm","wheelbase"], axis=1, inplace=True)
```

```python
df['CarName'] = df['CarName'].str.split(' ').str[0]
df['CarName'] = df['CarName'].str.lower()
```

```python
unique_values = df['CarName'].unique()
print(unique_values)
```

```python
df['CarNameId'] = pd.factorize(df['CarName'])[0]
df['drivewheelId'] = pd.factorize(df['drivewheel'])[0]
df['enginelocationId'] = pd.factorize(df['enginelocation'])[0]
df.head()
```

```python
word_to_num = {
    'zero': 0, 'one': 1, 'two': 2, 'three': 3, 'four': 4, 
    'five': 5, 'six': 6, 'seven': 7, 'eight': 8, 'nine': 9
}
df['cylindernumber'] = df['cylindernumber'].map(word_to_num)
```

```python
df.dropna(axis=0,inplace=True)
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap="viridis")
```

# Training Sets

```python
df.head()
```

```python
X = df[["CarNameId", "drivewheelId", "enginelocationId", "carlength","carwidth","curbweight","cylindernumber","enginesize","boreratio","stroke","horsepower","citympg","highwaympg"]]
y = df[["price"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.4, random_state=42)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression()
```

```python
linear_model.fit(X_train, y_train)
```

```python
print("Coef : ", linear_model.coef_)
```

```python
predictions = linear_model.predict(X_test)
```

```python
# Plot a graph that compares the y_test actual sample values
# to the result of the prediction from the fit using the same y_test samples
# A linear graph means the predictions are mostly correct
plt.scatter(y_test, predictions)
plt.xlabel("Y test")
plt.ylabel("Predicted Y")
```

```python
from sklearn import metrics

print("MAE", metrics.mean_absolute_error(y_test, predictions))
print("MSI", metrics.mean_squared_error(y_test, predictions))
# Square root of MSI
print("RMSE", np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
sns.displot((y_test-predictions), bins=100, kde=True)
```
