```python
word_to_num = {'B': 0, 'M': 1}
df['diagnosisId'] = df['diagnosis'].map(word_to_num)
df.drop(["Unnamed: 32"], axis=1, inplace=True)
df.drop(["id"], axis=1, inplace=True)
df.drop(columns=df.filter(like='_se').columns, inplace=True)
df.drop(columns=df.filter(like='_worst').columns, inplace=True)
df.head()
```

```python
df.info()
```

```python
sns.jointplot(x="fractal_dimension_mean", y="diagnosis", data=df)
```

```python
df.drop(["texture_mean","area_mean","smoothness_mean","concavity_mean","fractal_dimension_mean"], axis=1, inplace=True)
```

```python
df.head()
```

```python
# Find null (or none) values
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap="viridis")
```

```python
#model
# Drop null values
df.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[["radius_mean", "perimeter_mean", "compactness_mean", "concave points_mean","symmetry_mean"]]
y = df["diagnosis"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.4, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=5000)
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
print(predictions)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predictions))
```
